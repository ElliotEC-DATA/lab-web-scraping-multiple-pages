```python
def get_songs_info(songs_info):    
    songs_name = []
    autor_name = []
    duration_time = []
    music_style = []
    overall_rating = []
    links = []

    for song in song_content:
        
        songs = song.find("h3").get_text() if song.find("h3") else 'Not informed.'    
        authors = song.find('a', attrs= {'class' : 'Text__SCText-sc-kgt5u3-0 Link__PrimaryLink-sc-1vfcbn2-0 cBBiTW dPapvk'}).get_text(strip = True) if song.find('a', attrs= {'class' : 'Text__SCText-sc-kgt5u3-0 Link__PrimaryLink-sc-1vfcbn2-0 cBBiTW dPapvk'}) else 'Not informed.'
        duration = song.find('span', attrs = {'data-testid' : 'duration'}).get_text() if song.find('span', attrs = {'data-testid' : 'duration'}) else 'Not informed.'

        songs_name.append(songs)
        autor_name.append(authors)
        duration_time.append(duration)
    
    dct = {'Songs': songs_name, 'Author': autor_name, 'Duration': duration_time}
    
    return dct
```

```python
df = pd.DataFrame()

for start in starts:
    r = requests.get(f'https://www.senscritique.com/liste/mes_1000_chansons_preferees/2915989?page={starts}')
    soup = BeautifulSoup(r.content, 'html.parser')
    song_content = soup.find_all('div', attrs = {'class' : 'ProductListItem__Container-sc-ico7lo-0 hZLWXB'})
    info_dct = get_songs_info(song_content)
    new_df = pd.DataFrame.from_dict(info_dct)
    df = pd.concat([df, new_df])

df.insert(0, 'Rank', range(1, len(df)+1))
df
```

```python
df[['Songs', 'release date']] = df['Songs'].str.extract(r'^(.*?) \((\d+)\)$') # thanks GPT for this one
df
```

```python
def is_your_song_in_the_top(df):
    print('Is your song in the list ? Lets check ! ')
    author = input('The author of the song : ')
    title = input('The title of the song : ')
    if (df['Author'] == author).any() and (df['Songs'] == title).any():
        return "Yes, your song is in the top"
    else:
        return "No, your song is not in the top"
```
